```python
import numpy as np
import pandas as pd
from numpy import fromfile, dtype
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

# 1.The data file contains numerical attributes that describe a letter and its corresponding class. Read the datafile “letterCG.data” and set all the numerical attributes as features. Split the data in to train and test sets.

```python

binary_file = "c:/edureka/letterCG.bin"
dt = dtype([('Class', 'S1'), ('x-box', 'int32'), ('y-box', 'int32'),('width', 'int32'),('high', 'int32'),
            ('onpix', 'int32'),('x-bar', 'int32'),('y-bar', 'int32'),('x2bar', 'int32'),('y2bar', 'int32'),('xybar', 'int32'),
           ('x2ybr', 'int32'),(' xy2br', 'int32'),('x-ege', 'int32'),('xegvy', 'int32'),('y-ege', 'int32'),('yegvx ', 'int32')])
               
my_array = np.fromfile(binary_file,dt)
df = pd.DataFrame(my_array)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencode=LabelEncoder()
df.iloc[:,0]=labelencode.fit_transform(df.iloc[:,0])
df
```

```python
#Create an empty list called traget
target = []
for i in df['Class']:
    if i >= 0 and i <= 6:
        target.append('1')
    elif i >= 7 and i <= 13:
        target.append('2')
df['Target'] = target
```

```python
df.drop(['Class'],axis=1,inplace =True)
df.shape
```

```python
X= df.iloc[:,:16]
y = df['Target']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# Split the data in to train and test sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30,random_state=1)
```

# 2. Fit a sequence of AdaBoostClassifier with varying number of weak learners ranging from 1 to 16, keeping the max_depth as 1. Plot the accuracy on test set against the number of weak learners. Use decision tree classifier as the base classifier

```python
from sklearn.metrics import accuracy_score
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier #For Classification
from sklearn.ensemble import AdaBoostRegressor #For Regression
from sklearn.tree import DecisionTreeClassifier
```

# Using decision tree classifier as the base classifier and keeping the max_depth as 1.

```python
dt = DecisionTreeClassifier(max_depth=1,min_samples_split=16) 
abc = AdaBoostClassifier(n_estimators=50,base_estimator=dt,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
```

# Using decision tree classifier as the base classifier and keeping the max_depth as 2.

```python
dt = DecisionTreeClassifier(max_depth=2,min_samples_split=16) 
abc = AdaBoostClassifier(n_estimators=50,base_estimator=dt,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
```
