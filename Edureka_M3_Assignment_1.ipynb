# 3. Plotting: 

# Question 3 a. Make a bar plot  state name on the x -axis and their total benefitted inmates as their bar heights.Which state has the maximum number of beneficiaries?

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=df['STATE/UT']
y=df['total_benefitted']




plt.figure(figsize=(15,10))
plt.bar(x,y,width=2)

plt.xticks(x, x, rotation='vertical')
plt.xlabel('state name')
plt.ylabel('total benefitted inmates')
plt.title('maximum number of beneficiaries')
#plt.subplots_adjust(bottom=0.15)

plt.show()
```

# Question 3 b. pie chart that depicts the ratio among different modes of benefits.

```python
benefits=df.iloc[:, 2:6].sum()
benefits
```

```python
fig1, ax1 = plt.subplots()

activities = ['Inmates benefitted by Elementary Education','Inmates benefitted by Adult Education',
              'Inmates benefitted by Higher Education','Inmates benefitted by Computer Course']

colors = ['b', 'g','y','r']
plt.pie(benefits, labels=activities, colors=colors, startangle=90,autopct='%1.0f%%')
ax1.axis('equal')
plt.show()
```
